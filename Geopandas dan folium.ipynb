```python
import geopandas as gpd 
# memuat library geopandas pada python
```

```python
# Gunakan fungsi geopandas.read_file() untuk membaca shapefile
gdf = gpd.read_file("RBI25K_ADMINISTRASI_LN.geojson") 

# periksa tipe datanya sehingga dapat melihat bahwa bukan kerangka data normal, tetapi kerangka data GEOdataframe
gdf.head() 
# gdf.info()
```

```python
# pilih kolom yang diinginkan untuk petanya
# fillna (0) untuk menggantikan nilai null dalam kerangka data
gdf = gdf[['NAMOBJ', 'FCODE','REMARK','METADATA','geometry']].fillna(0)
```

```python
!pip install folium matplotlib mapclassify
# folium memvisualisasikan data yang sudah dimanipulasi dengan python dalam bentuk peta interaktif
# matplotlib library untuk membangun objek visual
# mapclassify skema klasifikasi peta
```

```python
import folium
m = gdf.explore(column="METADATA", name="MetaData") 
folium.LayerControl().add_to(m) 
m
#eksplorasi dan visualisasi data sebagai peta folium interaktif
```
